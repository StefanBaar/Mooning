```python
import glob

import numpy as np
import cv2
import matplotlib
from matplotlib import pylab as plt

import dehaze

from astropy.io import fits

from jupyterthemes import jtplot
jtplot.style(theme='onedork')

def to8bit(image):
    return cv2.cvtColor((image/image.max()*255).astype("uint8"),cv2.COLOR_GRAY2RGB)
def to8bitG(image):
    return (image/image.max()*255).astype("uint8")

def multi_Plot(data, COL = 1):
    fig, LL = plt.subplots(ncols=data.shape[0]/COL,nrows=COL, figsize=(15, 9),sharex = True, sharey = True)
    for i in range(len(LL)):
            L = LL.reshape(-1)
            L[i].imshow(data[i])
            L[i].grid(False)
            L[i].set_ylim([200,data.shape[1]-100])
    plt.show()
```

```python
im_list = glob.glob("*.fits")
moons = []
for i in im_list:
    moons.append(fits.open(i)[0].data)
moons = np.asarray(moons)
    
```

```python
fig, (L1,L2,L3) = plt.subplots(ncols=3, figsize=(15, 9))
L = [L1,L2,L3]
for i in range(len(L)):
    L[i].imshow(to8bit(moons[i])[:,:,0])
    L[i].grid(False)
    L[i].set_ylim([200,moons.shape[1]-100])
plt.show()
```

```python
badmoon = to8bit(moons[2].T)

def dehazing(badmoon, dark=45, tran=45, rec=0.4):
    I    = badmoon.astype('float64')/255.
    dark = dehaze.DarkChannel(I,dark);
    A    = dehaze.AtmLight(I,dark);
    te   = dehaze.TransmissionEstimate(I,A,tran);
    t    = dehaze.TransmissionRefine(badmoon,te);
    J    = dehaze.Recover(I,t,A,rec);
    return J

remoon = dehazing(badmoon)

plt.figure(figsize = (16,4))
plt.imshow(badmoon)
plt.grid(False)
plt.show()

plt.figure(figsize = (16,4))
plt.imshow(remoon)
plt.grid(False)
plt.show()
```

```python
 
def dehazeall(moons, dark=40, tran=100, rec=0.1):
    DH = []
    for i in moons:
        DH.append(to8bitG(i))
        DH.append(dehazing(to8bit(i), dark, tran, rec)[:,:,0])
    DH = np.array(DH)
    return DH

dehaze_data = dehazeall(moons, dark=1, tran=1, rec=0.1)
multi_Plot(dehaze_data, COL = 1)
```

```python
def get_list(data2D):
    x,y = np.where(data2D)
    return x,y

def linear_fit(x,y):
    return np.polyfit(x,y,1)

def outliar_mask(XX, s = 2.):
        limitU = XX.mean()+s*XX.std()
        limitD = XX.mean()-s*XX.std()
        return (XX < limitU) & (XX > limitD)
    
def apply_fit(x,y):
    A ,N = np.polyfit(x,y,1)
    Y    = A*x+N
    return Y


def plot_edge(data, COL=1):
    fig, LL = plt.subplots(ncols=data.shape[0]/COL,nrows=COL, figsize=(15, 9),sharex = True, sharey = True)
    for i in range(len(LL)):
        img = to8bitG(data[i])
        ret, thresh = cv2.threshold(img,1,1,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU) 
        canny = cv2.Canny(thresh,1,1,apertureSize = 3)
        y ,x  = get_list(canny)
        mask  = outliar_mask(y)
        x     = x[mask]
        y     = y[mask]
        
        yy = apply_fit(x,y)

        y2 = yy+(yy-y).max() 
        y1 = yy-(y-yy).min() 

        
        L = LL.reshape(-1)
        LL[i].imshow(img)
        LL[i].fill_between(x, y1, y2, color = "w" ,alpha = 0.3)
        LL[i].plot(        x ,y , "r.", label = "edge")
        LL[i].plot(        x,yy, "w", label = "lin fit")
        LL[i].set_ylim([200,moons[0].shape[0]-200])
        LL[i].legend()
    plt.show()
    
plot_edge(dehaze_data)
```

## Clustering 

```python
from sklearn.feature_extraction.image import grid_to_graph
from sklearn.cluster import AgglomerativeClustering
```

```python
def clustering(image, clusters = 2):
    X            = np.reshape(image, (-1,1))
    connectivity = grid_to_graph(*image.shape)
    ward         = AgglomerativeClustering(n_clusters    = clusters,
                                           linkage     = 'ward',
                                           connectivity= connectivity)
    ward.fit(X)
    return np.reshape(ward.labels_, image.shape)
```

```python
ALGO = []
for i in moons:
    ALGO.append(i)
    ALGO.append(clustering(i))
ALGO = np.asarray(ALGO)

multi_Plot(ALGO)
```
